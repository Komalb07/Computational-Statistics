#### 2.a)

```r
#set seed value
set.seed(436)

#Define all the required varaibles
r = 2
req_points = 200
count = 0
iterations = 0
points = matrix(NA, ncol=2, nrow=req_points)
 
while(count < req_points){
    #Generate coordinates x & y within the bounding box (square in this case)
    x = runif(n=1, min=-r, max=r)
    y = runif(n=1, min=-r, max=r)
    if (x^2+y^2<=r^2){
        count = count+1
        points[count, ] = c(x,y)
    }
    iterations = iterations + 1
}

cat("No. of iterations to make 200 succesful draws: ", iterations, "\n")

# Plotting all the generated points
plot(points[,1], points[,2], xlab="X-axis", ylab="Y-axis", main="Points inside the circle", col="red", asp=1)
```

#### 2.b)

```r
#set seed value
set.seed(436)

#Define all the required varaibles
r = 2
req_points = 200
count = 0
iterations = 0
points = matrix(NA, ncol=3, nrow=req_points)
 
while(count < req_points){
    #Generate coordinates x & y within the bounding box (cube in this case)
    x = runif(n=1, min=-r, max=r)
    y = runif(n=1, min=-r, max=r)
    z = runif(n=1, min=-r, max=r)
    if (x^2+y^2+z^2<=r^2){
        count = count+1
        points[count, ] = c(x,y,z)
    }
    iterations = iterations + 1
}

cat("No. of iterations to make 200 succesful draws: ", iterations, "\n")

# Plotting all the generated points
# This is a 3D scenario, so let's plot 2D projections ignoring each axis at a time
par(mfrow = c(2, 2))

plot(points[,1], points[,2], xlab="X-axis", ylab="Y-axis", main="Points inside the circle", col="red", asp=1)
plot(points[,1], points[,3], xlab="X-axis", ylab="Z-axis", main="Points inside the circle", col="red", asp=1)
plot(points[,2], points[,3], xlab="Y-axis", ylab="Z-axis", main="Points inside the circle", col="red", asp=1)

par(mfrow = c(1, 1))
```

We can notice a significant increase in the number of iterations to create a d-sphere as d increased from 2 to 3. **Explain in Video**   Use an appropriate method introduced in class to draw 10,000 rvs from this target distribution:

$$
\pi(x) = \frac{3}{4}(1-x^2), \text{ for } |x| \leq 1
$$

Also, make a histogram density estimate with the actual target $\pi(x)$ drawn over it for comparison.  Also calculate the ratio of acceptances to total number of draws. As always, any math and other statistical reasonin used should be explained!


Hint. Output should look like this (you don't need to make the curve blue though):


```r
#Set seed
set.seed(283)

#Define X and u
X = runif(15000, min=-1, max=1)
U = runif(15000, min=0, max=1)

#Define the target function
pi_x = function(x){
    new_val = (3/4)*(1-x^2)
}

# Define proposal function: In this case the proposal function is uniform between -1 and 1, 
# so density of the proposal function would be equal to 1/b-a = 1/1-(-1) => 1/2
g_x = 1/2

#The highest value of the target function is equal to 3/4 when x=0
# So the scaling factor c would be equal to 3/2 since, g_x = 1/2
c = 3/2

count = 1
accepted_samples = rep(NA, times=10000)
idx = 1

#Loop until we have either 15000 iterations or 10000 accepted samples
while(count<=15000 & idx<10000){
    #Retrieve the current random number from U
    test_u = U[count]

    # Calculate the threshold value for acceptance
    # pi_x(X[count]) represents the target distribution at X[count]
    # c * g_x represents the proposal distribution scaled by the constant c
    threshold = pi_x(X[count])/ (c*g_x)

    # If condition is satisfied, accept the sample X[count]
    if (test_u <= threshold){
        accepted_samples[idx] = X[count]
        idx = idx+1
    }
    count = count+1
    
}
#Ratio of acceptance to the total no.of draws
ratio = length(accepted_samples)/count
cat("Ratio:", ratio)

#Plot histogram of accepted sampels
hist(accepted_samples, probability=TRUE, breaks=25, ylim=c(0,0.8))
curve(pi_x, from=-1, to=1, add=TRUE, col='red', lwd=3)
```

The histogram of the accepted samples perfectly aligns with the target density pi_x. 
So, we can say that the accepted samples follows pi(x)

### Exercise 4:  Sample from Trapezoid

Recall the question from class where we have five students sitting at the coordinates:

(0.1, 0.8), (0.2, 0.9), (0.4, 0.3), (0.5, 0.5), (0.6, 0.7)



Now imagine that the room is a trapezoidal classroom where the four corners are at the points
(0,0), (0,1), (2,1), and (1.3, 0). The main idea here is to figure out how to sample from this irregularly shaped room.


 a) **Explain in Video** Similar to HW2, using simulation, please approximate the p-value that we would see data as extreme as this data (two-sided test using the same distance statistic as before) if the students really were sitting randomly, independently uniformly around the classroom. Make sure you justify which direction you use when for the inequality to get the pvalue.
 
 Hint: there might be other ways to generate this, but if you do what I did, then your first randomly generated statistic should be `7.71213031970951`. 

b) **Explain in Video**  Then answer: what is your conclusion about how students sat in class? Use hypothesis testing at the $\alpha=.05$ level.


```r
# Function to calculate the total pairwise distance between each pair of students
set.seed(434)
calculate_pairwise_distance <- function(xVec, yVec) {
  n <- length(xVec)
  total_distance <- 0
  for (i in 1:n) {
    for (j in 1:n) {
      # Calculate the Euclidean distance between each pair of points
      distance <- sqrt((xVec[i] - xVec[j])^2 + (yVec[i] - yVec[j])^2)
      total_distance <- total_distance + distance
    }
  }
  return(total_distance)
}

# Function to check if a point (x, y) is inside the trapezoid
is_point_in_trapezoid <- function(x, y) {
  # Condition 1: Check if point is in the rectangle or on the top line
  condition_1 <- (x >= 0 && x <= 1.3 && y >= 0 && y <= 1)
  
  # Condition 2: Check if point is inside the triangle
  condition_2 <- (x > 1.3  && y >= (1.4286 * x - 1.857) )
  
  # Return TRUE if either condition is met
  return (condition_1 || condition_2)
}

# Function to simulate random seating of students within the trapezoid
generate_random_seating <- function(num_students) {
  studentX <- numeric(num_students)
  studentY <- numeric(num_students)
  for (i in 1:num_students) {
    while (TRUE) {
      x <- runif(1, 0, 2)  # Random x between 0 and 2
      y <- runif(1, 0, 1)  # Random y between 0 and 1
      if (is_point_in_trapezoid(x, y)) {  # Check if point is valid
        studentX[i] <- x
        studentY[i] <- y
        break;
      }
    }
  }
  return(list(studentX = studentX, studentY = studentY))
}

# Function to run the simulation and calculate the p-value
run_simulation <- function(num_students, runs, data_stat) {
  saved_distances <- numeric(runs)
  
  for (r in 1:runs) {
    # Generate random seating arrangement
    seating <- generate_random_seating(num_students)
    studentX <- seating$studentX
    studentY <- seating$studentY
    
    # Calculate the pairwise distance for the simulated seating
    saved_distances[r] <- calculate_pairwise_distance(studentX, studentY)
  }
  
  # Calculate two-tailed p-value
  count_less <- sum(saved_distances <= data_stat)
  p_value <- (2 * count_less) / runs
  
  return(list(p_value = p_value, saved_distances = saved_distances))
}

# Main program starts here

# Coordinates of actual students in the classroom
xcoord <- c(0.1, 0.2, 0.4, 0.5, 0.6)
ycoord <- c(0.8, 0.9, 0.3, 0.5, 0.7)

# Calculate the pairwise distance for the given seating (data statistic)
data_stat <- calculate_pairwise_distance(xcoord, ycoord)
cat("The observed data statistic (total pairwise distance) is", data_stat, "\n")

# Simulation parameters
num_students <- 5
runs <- 1000

# Run the simulation and get the p-value
result <- run_simulation(num_students, runs, data_stat)
p_value <- result$p_value
saved_distances <- result$saved_distances

# Output the p-value
cat("The Two-Tailed Approximate P-value is:", p_value, "\n")
```
